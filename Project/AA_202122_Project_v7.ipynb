### Data Fields

* **track** - The Name of the track.
* **artist** - The Name of the Artist.
* **danceability** - Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable. 
* **energy** - Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low on the scale. Perceptual features contributing to this attribute include dynamic range, perceived loudness, timbre, onset rate, and general entropy. 
* **key** - The estimated overall key of the track. Integers map to pitches using standard Pitch Class notation. E.g. 0 = C, 1 = C?/D?, 2 = D, and so on. If no key was detected, the value is -1.
* **loudness** - The overall loudness of a track in decibels (dB). Loudness values are averaged across the entire track and are useful for comparing relative loudness of tracks. Loudness is the quality of a sound that is the primary psychological correlate of physical strength (amplitude). Values typical range between -60 and 0 db. 
* **mode** - Mode indicates the modality (major or minor) of a track, the type of scale from which its melodic content is derived. Major is represented by 1 and minor is 0.
* **speechiness** - Speechiness detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks. 
* **acousticness** - A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic. The distribution of values for this feature look like this:
* **instrumentalness** - Predicts whether a track contains no vocals. “Ooh” and “aah” sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly “vocal”. The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content. Values above 0.5 are intended to represent instrumental tracks, but confidence is higher as the value approaches 1.0. The distribution of values for this feature look like this:
* **liveness** - Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live.
* **valence** - A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).
* **tempo** - The overall estimated tempo of a track in beats per minute (BPM). In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration. 
* **duration_ms** - The duration of the track in milliseconds.
* **time_signature** - An estimated overall time signature of a track. The time signature (meter) is a notational convention to specify how many beats are in each bar (or measure).
* **chorus_hit** - This the the author's best estimate of when the chorus would start for the track. Its the timestamp of the start of the third section of the track.
* **sections** - The number of sections the particular track has. This feature was extracted from the data received by the API call for Audio Analysis of that particular track.
* **decade** - The decade in which the track was released. It can take on the following values: '60s', '70s', '80s', '90s', '00s', '10s'.

As a result of the analysis made of the EDA section. The first step of the preprocessing will be the creation of the feature *hit_artist*. The goal with the creation of this feature it is to translate the possible importance of the artists might have in predicting hit or flop. However, due to large amount of artists on the dataset it is not possible to apply one hot encoding. Label encoding, could be a possibility. Nonetheless, it is not possible to stablish an ordinal order between artists. Thus, the solution will be the binning through the number of hits and flops assign to each artist.

```python
"""computes the number of bins based on Rice's rule
formula is given by the cube_root(n)*2, where n is the size of dataset
https://www.statisticshowto.com/choose-bin-sizes-statistics/
Arguments: df  = dataframe that contains the feature to be binned"""
def compute_number_bins(df):
    n = len(df.index)
    return np.cbrt(n)*2
```

In order to perform the binning of the number of hits and flops of each artist. It is needed to define the number of bins to be used. Therefore, it was applied the Rice’s rule, where it did achieve better results than the Sturge’s Rule.

```python
"""does the binning for each feature.
Arguments: df = dataframe to perform binning
           n_bins = number of bins to be created
           name_bin = name of the column that will hold the bins interval"""
def bin_feature(df,feature,name_bin,n_bins=0,intervals = None):
    if np.any(intervals):
        if df[feature].min() not in intervals:
            intervals = np.insert(intervals,0,df[feature].min())
        if df[feature].max() not in intervals:
            intervals = np.append(intervals,df[feature].max())
        df[name_bin] = pd.cut(df[feature],bins = intervals) 
    else:
        df[name_bin] = pd.cut(df[feature],bins = n_bins)
        
```

```python
"""computes the mean for each bin created.
Arguments: df = dataframe to perform binning
           bin_feature = bin to be computed the mean"""
def compute_bin_mean(df,bin_feature):
    return df.groupby(df[bin_feature]).mean()
```

```python
"""for each value of the feature belonging to a bin assign the new mean of the bin.
Arguments: df = dataframe to perform binning
           bins_means = dataframe that holds the means corresponding to each bin
           feature = feature to be assign the mean of its value
           bin_feature = the corresponding bin that the value belongs"""
def assign_mean(df,bins_means,feature,bin_feature):
    for i, val in enumerate(bins_means.index):
        if val in df[bin_feature].values:
            #print(bins_means.loc[val,feature])
            df.loc[df.index[np.where(df[bin_feature].values == val)],feature] = bins_means.loc[val,feature]
```

```python
""" re-arranges the dataframe according to the list of columns passed
Arguments:df = df to be re-arranged
         columns_list = list of names of columns"""
def reindex_columns(df,columns_list):
    df = df.reindex(columns=columns_list)
    return df
```

After definning several functions that will help on the creation of the feature *hit_artist* it is defined the *create_hit_artist_feature* function that will perform the main computations and steps to create the feature.

```python
"""Creates feature hit_artist
Arguments: hit_artists_df = corresponding dataframe with the computed number of hits and flops
           df = corresponding dataframe"""
def create_hit_artist_feature(hit_artists_df,df):
    #get number of bins
    n_bins = round(compute_number_bins(hit_artists_df['n_hit']))
    copy_df = hit_artists_df.copy()
    #Perform the binning for features n_hit and n_flop
    for i, col in enumerate(hit_artists_df.columns[1:]):
        #Compute the bins
        bin_feature(copy_df,col,'bin_{}'.format(col),n_bins=n_bins)
        #Compute the mean of each bin
        bins_means_hit = compute_bin_mean(copy_df,'bin_{}'.format(col))
        #Assign the mean computed to each artist
        assign_mean(copy_df,bins_means_hit,col,'bin_{}'.format(col))
    #Compute the probability of hit after binning
    for i, val in enumerate(copy_df.index):
        hit = copy_df.iloc[i]['n_hit']
        flop = copy_df.iloc[i]['n_flop']
        copy_df.loc[i,'hit_artist'] = round(hit / (hit + flop),3)
        #print(val,copy_df.loc[val,'hit_artist'])
    #Merge the hit_artist feature within the original dataframe
    for i, val in enumerate(copy_df['artist']):
        df.loc[df.index[np.where(val == df['artist'].values)],'hit_artist'] = copy_df.loc[i,'hit_artist']
    #Reorganize dataframe
    column_names = ['track', 'artist', 'uri','hit_artist','danceability', 'energy', 'key', 'loudness',
       'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'duration_ms', 'time_signature', 'chorus_hit',
       'sections', 'decade', 'hit']
    df = reindex_columns(df,column_names)
    return df
```

Therefore, the creation of the *'hit_artist'* feature begins with the binning of the *n_hits* and *n_flops*. 

Having computed the bins and assigned the mean of those to the corresponding artist within the bin. It is computed the hit probability of each bin related to each artist. The results obtained are the new feature created named *'hit_artist'*.

After the newly created feature, *'hit_artist'*, it is merged with the original dataframe and the dataframe was reindexed, in order to maintain the last column as the target feature.

```python
"""computes the bins for each feature. Where the bounds of each bins corresponds 
    to the splits nodes defined by a decision tree
    Arguments: bin_df = dataframe to be performed the bins 
               depth = specify depth limit a tree should try to grow"""
def binning(df_bin, depth=20, optimize = True):
    #Obtain x and y to train and test the decision tree
    X_train, X_test, y_train, y_test = train_test_split(df_bin[df_bin.columns[:-1]],df_bin[df_bin.columns[-1]])
    train_scores = []
    
    #Loop trough to each feature to perform the binning
    for i, feature in enumerate(X_train.columns):
        if optimize:
            #Optimize hyperparameter max_depth of tree
            for j in range(1,depth):
                dtc_model = DecisionTreeClassifier(max_depth=int(j))
                scores = cross_val_score(dtc_model,X_train[feature].to_frame(),y_train,cv=5)
                train_scores.append(scores.mean())
            #Fit tree based on optimal_max_depth
            optimal_max_depth = np.argmax(train_scores) + 1
            dtc_model = DecisionTreeClassifier(max_depth=optimal_max_depth, random_state=42).fit(X_train[feature].to_frame(),y_train)
        else:
            dtc_model = DecisionTreeClassifier(max_depth=depth, random_state=42).fit(X_train[feature].to_frame(),y_train)
        #Solution of thresholding from: 
        #https://stackoverflow.com/questions/44644945/binning-data-via-decisiontreeclassifier-sklearn
        #Get the split nodes to determine the intervals 
        threshold = dtc_model.tree_.threshold[dtc_model.tree_.threshold > -2]
        threshold = np.sort(threshold)
          
        feature_name = '{}_bin'.format(feature)
        #Bin the value of features
        bin_feature(df_bin,feature,feature_name,intervals=threshold)
        #Compute the mean of each bin
        bins_means = compute_bin_mean(df_bin,feature_name)
        #Attribute the mean of each bin to the values within the bin
        assign_mean(df_bin,bins_means,feature,feature_name)
    #Remove columns indicating the intervals 
    for i, val in enumerate(df_bin.columns):
        if val.endswith('bin'):
            df_bin = df_bin.drop(columns=[val])
    return df_bin
    
```

The definition of the bins for each feature is based on the split nodes generated by a decision tree trained only with each continuous feature. By applying this method, it is assured that the splitting points are optimal. Thus, the values within each bin will be more homogeneous within the interval  than to the values belonging to another bin [2].

Before performing the binning the features need to be normalized. This will allow to compute the bins and it will improve the performance of further models, specially the linear models. Thus it is applied to the two dataframes the min-max transformation, that scales all the values to the interval range of [0,1]. 

```python
"""normalizes the values into a range [0,1].
Arguments: df = dataframe to be normalized
           features = features to be normalized"""
def normalize_features(df,features):
    #Store categorical features to avoid normalize it
    categorical = ['track','artist','uri','key','mode','time_signature','decade']
    #Normalize features
    for i, feature in enumerate(features):
        if feature not in categorical:
            min_val = df[feature].min()
            max_val = df[feature].max()
            den = max_val - min_val
            #Check if feature is already normalized
            if abs(den) > 1:
                df[feature] = (df[feature] - min_val) / den
```

Moreover, the feature *decade* needed to be encoded. This variable can be thought as an ordinal feature.

```python
"""it encodes decade by ordinal order
Arguments: df = dataframe to be performed the encode"""
def ordinal_encode_decade(df):
    #Define ordinal labels
    labels_map = {"60s":0,"70s":1,"80s":2,"90s":3,"00s":4,"10s":5}
    #perform encoding
    df['decade'] = df['decade'].replace(labels_map)
    return df['decade']
```

```python
"""puts hit column as last column
Arguments: df = dataframe"""
def reindex_hit(df):
    hit_col =df['hit']
    df = df.drop(columns=['hit'])
    df.insert(len(df.columns),'hit',hit_col)
    return df
```

```python
"""performs the one-hot-encode for a passed featured.
Arguments: df = dataframe to perform one-hot-encode
           name_dummy = feature to be encoded"""
def get_dummy(df,feature,name_dummy):
    dummy = pd.get_dummies(df[feature], prefix = name_dummy)
    df = pd.merge(left = df, right = dummy, left_index=True,
        right_index=True)
    del df[feature]
    return df
```

```python
"""main function to compute and perform the preprocessing steps definef
Arguments: to_bin = bool value to indicate if binning is to be performed"""
def preprocessing_data(to_bin=False):
    #Load the dataset and convert to dataframe
    df = load_dataset("Spotify_Dataset.csv")
    #Create hit_artist feature
    df = create_hit_artist_feature(hit_artists_df,df)
    #Normalize features
    normalize_features(df ,df.columns)
    #Check for binning
    if to_bin:
        #Perform binning
        features_to_bin = ['danceability', 'energy', 'loudness','speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'duration_ms', 'chorus_hit','hit']
        df[features_to_bin] = binning(df[features_to_bin].copy())
    
    #One hot encode categorical features
    dummy_features = ['mode','key','time_signature']
    for i, feature in enumerate(dummy_features):
        df = get_dummy(df,feature,'{}_'.format(feature))
    #Ordinal encode
    df['decade'] = ordinal_encode_decade(df)
    #Reindex hit column
    df = reindex_hit(df)
    
    return df
```

```python
"""performs the optimization of the parameter(s) of a given model.
Arguments: model (lasso, logreg)  = specify the name of the model, currently supporting (Lasso, Logistic Regression)
           X = feature values
           y = class values
           pipeline = Pipeline to be passed, default None"""
def optimize_hyperparameter(model,X,y,pipe_list=None):
    #list to store scores and hyperparameter values
    train_scores = []
    #Optimize Logistic Regression
    if model == 'logreg':
        #Define array of values for C
        C_list = np.linspace(0.01, 10, 100)
        #Check for pipeline
        if pipe_list:
            #Loop through array of values for C
            for i, val in enumerate(C_list):
                #Define pipeline to use with resampling method
                pipeline = Pipeline([pipe_list[0],('logreg',LogisticRegression(C=val, solver='liblinear'))])
                #Perform cross-validation
                scores = cross_val_score(pipeline,X,y,cv=5)
                #Store score and hyperparameter values
                train_scores.append({'mean score':scores.mean(), 'C': val})
        else:
            #Loop through values for C
            for i, val in enumerate(C_list):
                #Define model with the value for C
                logreg = LogisticRegression(C=val, solver='liblinear')
                #Perform cross-validation
                scores = cross_val_score(logreg,X,y,cv=5)
                #Store score and hyperparameter values
                train_scores.append({'mean score':scores.mean(), 'C': val})
        #Convert to dataframe
        train_scores = pd.DataFrame(train_scores)
        #Return value for C that obtained the biggest mean score and dataframe of train scores
        return C_list[np.argmax(train_scores['mean score'].values)], train_scores
    #Optimize SVC
    elif model == 'svc':
        #Declare SVC
        svc = SVC(kernel='rbf')
        #Check for pipeline
        if pipe_list:
            #Define pipeline with resampling method
            pipeline = Pipeline([pipe_list[0],('svc',svc)])
            #Define grid of values to be performed gridsearch
            param_grid = { 'svc__C': np.linspace(50,100,2),
                    'svc__gamma':np.linspace(0.01,0.1,2) }
            #Perform GridSearch and find optimal C and gamma
            grid= GridSearchCV(pipeline,param_grid)
            grid.fit(X_train_valence, y_train_valence)
            optimal_C = grid.best_params_['svc__C']
            optimal_gamma = grid.best_params_['svc__gamma']
            train_scores = grid.cv_results_['mean_test_score']
        else:
           #Define grid of values to be performed gridsearch
            param_grid = { 'C':[100],
                    'gamma':np.linspace(0.01,0.1,2) }
            #Perform GridSearch and find optimal C and gamma
            grid= GridSearchCV(svc,param_grid)
            grid.fit(X, y)
            optimal_C = grid.best_params_['C']
            optimal_gamma = grid.best_params_['gamma']
            train_scores = grid.cv_results_['mean_test_score']
        #Return optimal C, gamma and train_Scores
        return optimal_C, optimal_gamma, train_scores
    #Optimize knn
    elif model == 'knn':
        if pipe_list:
            for k in range(1,20):
                pipeline = Pipeline([(pipe_list[0]),('knn',KNeighborsClassifier(n_neighbors=k))])
                scores = cross_val_score(pipeline,X,y,cv=5)
                train_scores.append({'mean score':scores.mean(), 'n_neighbors': k})
        else:
            for k in range(1,20):
                knn = KNeighborsClassifier(n_neighbors=k)
                scores = cross_val_score(knn,X,y,cv=5)
                train_scores.append({'mean score':scores.mean(), 'n_neighbors': k})
        
        train_scores = pd.DataFrame(train_scores)
        return np.argmax(train_scores['mean score'].values), train_scores
    #Optimize Decision trees
    elif model == 'dct':
        if pipe_list:
             for i in range(1,26):
                pipeline = Pipeline([(pipe_list[0]),('dct',DecisionTreeClassifier(max_depth=i))])
                scores = cross_val_score(pipeline,X,y,cv=5)
                train_scores.append({'mean score':scores.mean(), 'max_depth': i})
        else:
            for i in range(1,26):
                dtc_model = DecisionTreeClassifier(max_depth=i)
                scores = cross_val_score(dtc_model,X,y,cv=5)
                train_scores.append({'mean score':scores.mean(), 'max_depth': i})
        train_scores = pd.DataFrame(train_scores)
        return np.argmax(train_scores['mean score'].values), train_scores
          
```

In order to analyze the assigned coefficients for each feature, it needed to understand briefly how the Logistic Regression works. A Logistic model tries to establish a relationship between the target variable and the features. By transforming the probability of an event, over the probability of the complement, into odds. Then, applying a log transformation, will make the values to range from negative infinity to positive infinity. To these transformations are called the logit transformation and they are monotonic transformations. Therefore, the Logistic transformation models this logit transformed probability in a linear relation [5].

Thus, the coefficient values assigned to each feature can be interpreted in terms of importance in the occurrence of hit or flop. Features attributed a greater weight and with positive sign, are likely to have more impact on the occurrence of a hit. Whether the opposite can be acknowledged for features with greater negative weights. That are more prone to the occurrence of flop. Hence, features with coefficients close to 0 are regarded as having low importance when predicting a hit or flop.

On this case, the Logistic Regression model achieved a training score of approximately 0.74. When looking for the coefficients assigned to each feature by the Logistic Regression, it is noticed the apparently value of 0 assigned to the feature *’time_signature__0’*. Although, when applying Logistic Regression, it is not possible to reach a value 0 for the coefficient. Therefore, that should be analyzed as value very close to 0, that it is impossible to represent in computer. Emphasizing the low importance of this feature. Furthermore, for most of the coefficients relating to the *’key’*, except *’key__2’*, liveness, valence and decade were assigned a low score, near of 0. 

On the other hand, the coefficients of the features like *’danceability’* and *’loudness’* were attributed high positive values. Thus, these are features with high significance on the occurrence of hit. Likewise, to the coefficients *’speechiness’* and *’instrumentalness’* were attributed bigger weights with negative sign. Hence, they have a great impact on the occurrence of a flop.

It is defined several functions that will be called at the time of training and test the classifiers. 

```python
"""computes the measures for a binary and multi-classifier
Arguments: y_true = corresponds to the y_test
           y_pred = correspond to the output of predict method"""
def compute_metrics(y_true,y_pred):
    #Declare dict to store the measures
    metrics = {}
    #Initialize train_accuracy
    metrics['train_accuracy'] = -1
    #Compute test accuracy
    metrics['test_accuracy'] = accuracy_score(y_true,y_pred)
    #Compute the error
    metrics['error'] = 1 - metrics['test_accuracy']
    #Check if it is binary or multi class
    if len(np.unique(y_true)) > 2:
        #For multi classe compute the precision, recall and f_measure based on average weighted
        metrics['precision'] = precision_score(y_true,y_pred,average='weighted')
        metrics['recall'] = recall_score(y_true,y_pred,average='weighted')
        metrics['f_measure'] = f1_score(y_true,y_pred,average='weighted')
    else:
        #Compute precision, recall and f_measure
        metrics['precision'] = precision_score(y_true,y_pred)
        metrics['recall'] = recall_score(y_true,y_pred)
        metrics['f_measure'] = f1_score(y_true,y_pred)
    return metrics
```

```python
""" computes confusion matrix for the model
Arguments: clf = classifier
           X_train, X_test, y_train, y_test = Corresponding training and test sets"""
def compute_confusion_matrix_measures(clf,X_train,X_test,y_train,y_test):
    #Predict
    y_pred = clf.fit(X_train,y_train).predict(X_test)
    #Compute the performance measures of the model
    metrics = compute_metrics(y_test,y_pred)
    #Compute the confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    return cm,metrics
```

```python
"""plots the confusion matrix/matrices
Arguments: cm_list = list of confusion matrices
           n_rows = number of rows needed to display all confusion matrices, default (1)
           n_cm = number of confusion matrices to be plotted
           labels = labels of the classes
           titles = list of titles to be displayed"""
def plot_confusion_matrix(cm_list,labels,titles,n_rows=1):
    #Plot confusion matrices
    if len(cm_list) > 1:
        #Define fig and ax
        fig,ax = plt.subplots(n_rows,len(cm_list), figsize=(10,10))
        fig.tight_layout(w_pad=10)
        #Loop through list of confusion matrices and plot each
        for i in range(len(cm_list)):
            sns.heatmap(cm_list[i], square=True, annot=True, fmt='d', cbar=False,
                    xticklabels=labels,
                    yticklabels=labels, cmap='Blues', ax = ax[i])
        
            ax[i].set_xlabel('predicted label')
            ax[i].set_ylabel('true label')
            ax[i].set_title(titles[i], fontsize=11)
    #Plot confusion matrix
    else:
        cfmt_plot = sns.heatmap(cm_list[0], square=True, annot=True, fmt='d', cbar=False,
                    xticklabels=labels,
                    yticklabels=labels, cmap='Blues')
        cfmt_plot.axes.set_xlabel('predicted label')
        cfmt_plot.axes.set_ylabel('true label')
        cfmt_plot.axes.set_title(titles[0], fontsize=10)
    
    plt.show()
```

```python
"""fits the model and computes the performance measures and confusion matrix
Arguments: model: model to be fitted
           X_train_X_test,y_train,y_test = training and test set"""
def fit_model(model,X_train,X_test,y_train,y_test):
    model.fit(X_train,y_train)
    cm, mtcs = compute_confusion_matrix_measures(model,X_train,X_test,y_train,y_test)
    mtcs['train_accuracy'] = model.score(X_train,y_train)
    return cm,mtcs
```

```python
"""call the necessary functions for binary and multiclass classification
Arguments: model_dict = dict with models to be trained
           datasets = dict with the training and test sets
           labels = class labels
           """
#Adicionar train_test_split dentro da função passar dataset como table_X e table_y
def clf_train_test(models_dict,datasets,labels,pipeline=False):
    #Define list to store dataframe of the measures computed for each model
    metrics_models_df = []
    #Check if model is passed as pipeline
    if pipeline:
        for model in models_dict:
            for dataset in datasets:
                cm_list = []
                mtcs_list = []
                cm,mtcs = fit_model(models_dict[model],datasets[dataset][0],datasets[dataset][1],datasets[dataset][2],datasets[dataset][3])
                cm_list.append(cm)
                mtcs_list.append(mtcs)
            #plot the confusions metrics 
            titles = ['CM-Test Set {} {}'.format(key,model) for key in datasets]
            plot_confusion_matrix(cm_list,labels,titles)
            titles_df = ['{}'.format(key) for key in datasets]
            #print the dataframe measures
            metrics_df = pd.DataFrame(mtcs_list,index=titles_df)
            print('\033[1m' '{}'.format(model) + '\033[0m')
            print(metrics_df)
            print('\n')
            
    else:
        #Loop through the dict of models
        for model in models_dict:
            #Declare list to store confusions matrices and the dicts of measures
            cm_list = []
            mtcs_list = []
            #Loop through each datatset
            for i, dataset in enumerate(datasets):
                if model == 'gnb':
                    #fit the model
                    cm,mtcs = fit_model(models_dict[model],datasets[dataset][0],datasets[dataset][1],datasets[dataset][2],datasets[dataset][3])
                    cm_list.append(cm)
                    mtcs_list.append(mtcs)
                else:
                    #fit the model
                    cm,mtcs = fit_model(models_dict[model][i],datasets[dataset][0],datasets[dataset][1],datasets[dataset][2],datasets[dataset][3])
                    cm_list.append(cm)
                    mtcs_list.append(mtcs)

            #plot the confusions metrics 
            titles = ['CM-Test Set {} {}'.format(key,model) for key in datasets]
            plot_confusion_matrix(cm_list,labels,titles)
            titles_df = ['{}'.format(key) for key in datasets]
            #print the dataframe measures
            metrics_df = pd.DataFrame(mtcs_list,index=titles_df)
            print('\033[1m' '{}'.format(model) + '\033[0m')
            print(metrics_df)
            print('\n')
    
```

Before searching for the best resamoling method, it is required to perform the necessaries transformations to the dataset done previously, like enconding the target variable, scale the values of the dataset and split it into training and test set.

The function *find_optimal_sampling_method* will be used to further investigate the best resampling method to be applied. By computing the training accuracy scores.

```python
"""Train a classifier with different resampling methods
Arguments: X = train set 
           y = target train set
           clf = classifier to be trained"""
def find_optimal_sampling_method(X,y,clf):
    #Define seed of initialization for each method required
    seed = 42
    #Define dict of resampling methods
    methods = {
        'SMOTE': SMOTE(random_state = seed),
        'TomeKLinks': TomekLinks(),
        'ADASYN' : ADASYN(sampling_strategy='minority',random_state = seed),
        'SMOTETomek': SMOTETomek(random_state = seed),
    }
    #Declare empty list
    train_scores = []
    #Loop through dictionary of resampling methods
    for method in methods:
        #Define pipeline
        pipeline = make_pipeline(methods[method],clf)
        #Perform cross-validation
        scores = cross_val_score(pipeline,X,y,cv=5)
        #Append method and corresponding mean acc score
        train_scores.append({'resampling method':method,'acc score':scores.mean()})
    #Convert to Dataframe
    train_scores_df = pd.DataFrame(train_scores)
    #Select method that produced the biggest score
    best_method = train_scores_df.loc[train_scores_df['acc score'].idxmax(),'resampling method']
    return methods[best_method]
```

When dealing with imbalanced datasets, one of methods that can be applied are the resampling techniques. Within the resampling it can be applied strategies of under-sampling, over-sampling or combination of both. The under-sampling method consists of removing examples of the majority class(es). On the other hand, over-sampling refers to the technique of replicating examples of the minority class(es). Both techniques can compensate uneven datasets. Nonetheless, they entail trade-offs, as with the over-sampling overfitting might surge and with under-sampling, there is the loss of information [7].

Bearing in mind the trade-offs of the resampling methods, it was chosen 4 resampling techniques, of different categories. Thus, the <b>TomekLinks</b> which is an under-sampling method, that selects pair of examples based on the rules: Let a pair of examples be (x,y) if the nearest neighbor of x is y and vice-versa and both belong to different classes, minority and majority, so it is a Tomek Link. Then, the example belonging to the majority class is removed.  

Furthermore, it was selected the method <b>Synthetic Minority Oversampling Technique</b> (SMOTE) which is an over-sampling method. In this method, an example of the minority class is picked at random. Afterwards, it is proceeded to finding the k nearest neighbors of that data point and selects one of them randomly. Then a synthetic data point is generated between these two examples. Moreover, it was selected the <b>Adaptive Synthetic Sampling Approach</b> (ADASYN). Which is similar to the SMOTE but it uses an estimate of the local distribution of the minority class [8].

Additionally, SMOTETomek combines both SMOTE and Tomek, which is a combination of under-sampling and over-sampling. This technique aims to clean possible overlapping data points resulting of the over-sampling mitigating possible overfitting situation of the classifier [9].

```python
#best resampling method for Logistic Regression
logreg = LogisticRegression(C=10,solver='liblinear',max_iter=10000)
best_method_logreg = find_optimal_sampling_method(X_train_valence,y_train_valence,logreg)
#best resampling method for SVC
svc = SVC(C=100,kernel='rbf',gamma=0.1)
best_method_svc = find_optimal_sampling_method(X_train_valence,y_train_valence,svc)
#best resampling method for KNN
knn = KNeighborsClassifier(n_neighbors=13)
best_method_knn = find_optimal_sampling_method(X_train_valence,y_train_valence,knn)
#best resampling method for Naive Bayes
gnb = GaussianNB()
best_method_gnb = find_optimal_sampling_method(X_train_valence,y_train_valence,gnb)
#best resampling method for Decision Trees
dct = DecisionTreeClassifier(max_depth=9)
best_method_dct = find_optimal_sampling_method(X_train_valence,y_train_valence,dct)
```

<h2><a name="associations" style="color:black">2.3. Finding Associations</a></h2>

```python
"""computes the frequent itemsets
Arguments: binary_dataset = Binary dataset or sparmatrix
           min_supp = Threshold of minimum support to generate itemsets
           max_length = max length of the the frequent itemsets"""
def compute_itemsets(binary_dataset,min_supp,max_length):
    frequent_itemsets = apriori(binary_dataset, min_support=min_supp,use_colnames=True)
    frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
    frequent_itemsets = frequent_itemsets[(frequent_itemsets['length'] <= max_length)]
    return frequent_itemsets
```

When computing the frequent itemsets the apriori algorithm uses a minimum support threshold defined by the user. Thus, the support can be viewed as a simple proportion of the number of transactions in which a item appears over the total number of transactions [11].

Having computed the binary dataset, it is possible now to pass it to the apriori algorithm, in order to compute frequent item sets. Although, binning was performed enhancing the discretization of each feature. There is still a width variety of values for each feature, and being a large dataset, it will be defined minimum support of 5% and length 3. This will let to extract more rules from the dataset and not overflow the memory.

```python
freq_itemsets = compute_itemsets(matrix_items,0.05,3)
freq_itemsets
```

From the dataset with the support and length passed it was able to compute 574 frequent item sets. Which will now be used to extract rules. The extraction of rules will be focuses on finding logic expressions related to a hit and flop and discrete valence.



For the evaluation of the rules mined it will be given more attention to the measure’s **confidence** and **lift**. The confidence is simply the conditional probability of the *consequent* given the *antecedents*. Therefore, the confidence can be interpreted, in this case, as the likelihood of a song being a hit or flop given that some *antecedents* have occurred [12]. Furthermore, the lift can be viewed as a correlation measure, between the *consequent* and the *antecedents*. Thus, values higher 1 mean that the occurrence of those *antecedents* have a positive effect on the occurrence of the *consequent*. On the other hand, a lift lesser than 1 means the opposite. Finally, a lift near 1 can be interpreted as the *consequent* and *antecedents* are independent, although they appear very often together no correlation can be concluded. Thus, confidence and lift act as complementary measures that allow to perform a better association rule mining [13].

```python
"""Computes associative rules based on a given consequent and outputs the top n-th rules based on measure passed
Arguments: itemsets = frequent_itemsets
           threhold = min_threshold for confidence to extract rules
           conseq = desirred consequents to be extracted rules
           measure = measure to filter the returned rules
           n = number of rules to be returned"""
def find_rules(itemsets,threshold,measure,n,conseq = ['hit_0','hit_1']):
    rules = association_rules(itemsets,metric='confidence',min_threshold=threshold)
    rules['consequents']=rules['consequents'].apply(lambda x: "".join(list(x)))
    rules = rules.loc[rules['consequents'].isin(conseq)]
    return rules.sort_values(measure, ascending = False).head(n)
```

```python
"""plot features distributions
Arguments df = dataframe
          features = features to be plotted"""
def plot_distribution(df,features):
    for i, feature in enumerate(features):
        sns.histplot(data=df,x=feature, color ='orange')
        plt.show()
```

In order to perform the clustering, it will be used the K-means algorithm, which is a distance-based algorithm. It works as an iterative process, where at first are assigned at random centroids, then the instances will be labeled and the centroids updated. These steps will be performed until it converges for an optimal solution. However, the initialization  of the centroids might lead to sub-optimal convergences [14].
On the specifc case of the K-means implementation by sckit-learn, it provides as default initialization, called *k-means++*. Which is a smart way of initialize the cluster centroids, that will speed up the convergence [15].

Before proceeding with finding the clusters it is needed to find the best number of clusters, on the case of K-means that refers the hyperparametere k.
Therefore, it will be applied the silhouette coefficient. This metric is defined by (b-a)/max(a,b), where a is the mean distance to the other data points in the same cluster and b is the nearest-cluster distance. The silhouette coefficient can vary between -1 and +1. Hence, a result new of +1 it means that the instance is well inside its own cluster and far from the other clusters. Distance. Values near 0 indicate overlapping clusters. Finally, negative values present that a sample has been assigned to the wrong cluster, as a different cluster is more similar [16].

```python
#Solution from: https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html#sphx-glr-auto-examples-cluster-plot-kmeans-silhouette-analysis-py
"""plot silhouette diagrams
Arguments: X = dataframe.values
           k = number of clusters
           labels = labels of the clusters
           score = silhouette coefficient"""
def plot_silhouette(X,k,labels,score):

    # Create a subplot with 1 row and 2 columns
    fig = plt.figure(figsize=(5,5))

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    plt.xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    plt.ylim([0, len(X) + (k + 1) * 10])


    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, labels)

    y_lower = 10
    for i in range(k):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = sample_silhouette_values[labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / k)
        plt.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            ith_cluster_silhouette_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7,
        )

        # Label the silhouette plots with their cluster numbers at the middle
        plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    plt.title("The silhouette plot for the various clusters.")
    plt.xlabel("The silhouette coefficient values")
    plt.ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    plt.axvline(x=score, linestyle="--")

    plt.yticks([])  # Clear the yaxis labels / ticks
    plt.xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    plt.show()
```

```python
"""finds best number of cluster using k-means based on silhouette coefficient
Arguments: X = dataframe.values"""
def find_optimal_k(X):
    cluster_scores = []
    for k in range(2,8):
        kmeans = KMeans(n_clusters=k,random_state=0)
        labels = kmeans.fit_predict(X)
        score = {'score':silhouette_score(X,labels),'k':k}
        plot_silhouette(X,k,labels,score['score'])
        cluster_scores.append(score)
    return cluster_scores
```

```python
dct_cluster_hca = tree.DecisionTreeClassifier()
dct_cluster_hca = dct_cluster_hca.fit(cluster_df[cluster_df.columns[:-1]].values,cluster_df[cluster_df.columns[-1]].values)
dot_data = tree.export_graphviz(dct_cluster_hca, out_file="cluster_hca.dot",
                                feature_names= cluster_df.columns[:-1].values,
                                class_names=np.unique(labels_hca).astype(str),
                                filled=True, rounded=True,
                                special_characters=True)

(graph,) = pydot.graph_from_dot_file('cluster_hca.dot')
graph.write_png('cluster_hca.png')
img = mpimg.imread('cluster_hca.png')

plt.figure(figsize = (40,40))
imgplot = plt.imshow(img)
```

Oppositely, to the tree generated for the k-means, this tree is much more complex, which doesn’t facilitate the analysis and interpretation of each cluster.
